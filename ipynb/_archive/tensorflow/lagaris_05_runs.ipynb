```python
import matplotlib.pyplot as plt
import numpy as np
```

## Record the results of the grid of runs.

```python
eps_rel_converged = 1e-6
```

```python
n_hidden = np.array([10, 20, 40, 80, 160, 320, 640, 1280])
```

```python
n_xy_train = np.array([10, 20, 40, 80])
```

```python
epochs_converged = np.array([
    [2050, 2066, 860, 523, 213, 8362, 4282, 16000],
    [1725, 3044, 4381, 423, 266, 16000, 3173, 5837],
    [2811, 852, 323, 429, 244, 16000, 8874, 11570],
    [3011, 1152, 583, 421, 9081, 16000, 8331, 6526]
])
```

```python
time_converged = np.array([
    [35.270332, 36.10199, 14.614452, 9.638844, 3.674355, 146.277326, 77.832817, 359.305243],
    [29.621149, 52.396164, 75.715874, 8.480345, 4.737664, 343.986065, 78.876072, 170.077427],
    [54.516404, 15.291649, 5.904163, 8.992404, 5.896808, 497.243594, 389.957919, 764.206633],
    [61.451976, 28.996579, 16.205016, 14.220754, 443.53544, 1144.692577, 729.470055, 942.086825],
])
```

```python
time_per_epoch = time_converged/epochs_converged
```

```python
loss_converged = np.array([
    [0.018498, 0.018767, 0.025681, 0.029897, 0.032281, 0.018479, 0.028778, 0.018558],
    [0.014155, 0.014923, 0.017268, 0.027352, 0.029982, 0.012713, 0.046367, 0.024883],
    [0.008546, 0.020767, 0.02501, 0.026187, 0.028695, 0.011589, 0.01621, 0.014489],
    [0.00785, 0.019174, 0.020728, 0.025512, 0.012188, 0.011418, 0.01803, 0.024279]
])
```

```python
rmse_converged = np.array([
    [0.002584989, 0.002488197, 0.002653936, 0.002615775, 0.002719052, 0.002453838, 0.002666901, 0.002457891],
    [0.002602742, 0.002542171, 0.002591821, 0.002700918, 0.002715876, 0.002490769, 0.003130686, 0.002692814],
    [0.002510148, 0.002767494, 0.002695296, 0.002697566, 0.002790396, 0.002540275, 0.00257288, 0.002554647],
    [0.002540987, 0.002766316, 0.002724773, 0.002711341, 0.00261686, 0.002585448, 0.002660905, 0.002767667]
])
```

## Epochs to convergence

```python
for (i, y) in enumerate(epochs_converged):
    plt.plot(n_hidden, y, label="$n_{train}=%s$" % n_xy_train[i]**2)
plt.title("Epochs needed to achieve %s relative convergence" % eps_rel_converged)
plt.xlabel("Number of hidden nodes")
plt.ylabel("Epochs")
plt.legend()
plt.show()
```

## Time to convergence

```python
for (i, y) in enumerate(epochs_converged):
    plt.plot(n_hidden, y, label="$n_{train}=%s$" % n_xy_train[i]**2)
plt.title("Time needed to achieve %s relative convergence" % eps_rel_converged)
plt.xlabel("Number of hidden nodes")
plt.ylabel("Time (seconds)")
plt.legend()
plt.show()
```

## Time per epoch

```python
for (i, y) in enumerate(time_per_epoch):
    plt.plot(n_hidden, y, label="$n_{train}=%s$" % n_xy_train[i]**2)
plt.title("Time per epoch")
plt.xlabel("Number of hidden nodes")
plt.ylabel("Time (seconds) per epoch")
plt.legend()
plt.show()
```

## Loss function at convergence

```python
for (i, y) in enumerate(loss_converged):
    plt.plot(n_hidden, y, label="$n_{train}=%s$" % n_xy_train[i]**2)
plt.title("Loss function at convergence")
plt.xlabel("Number of hidden nodes")
plt.ylabel("Loss function")
plt.legend()
plt.show()
```

## RMS solution error at convergence

```python
for (i, y) in enumerate(rmse_converged):
    plt.plot(n_hidden, y, label="$n_{train}=%s$" % n_xy_train[i]**2)
plt.title("RMS solution error at convergence")
plt.xlabel("Number of hidden nodes")
plt.ylabel("RMS error")
plt.legend()
plt.show()
```
