```python
from math import exp
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def ya(x):
    return 1 + exp(-x**2 / 2)

def dyadx(x):
    return -x * exp(-x**2 / 2)

def d2yadx2(x):
    return exp(-x**2 / 2) * (x**2 - 1)
```

```python
x = np.arange(-1, 1, 0.01)
n = len(x)
y = np.zeros(n)
dy_dx = np.zeros(n)
d2y_dx2 = np.zeros(n)
```

```python
for i in range(n):
    y[i] = ya(x[i])
    dy_dx[i] = dyadx(x[i])
    d2y_dx2[i] = d2yadx2(x[i])
plt.plot(x, y, label = "$y$")
plt.plot(x, dy_dx, label = "$dy/dx$")
plt.plot(x, d2y_dx2, label = "$d^2y/dx^2$")
plt.xlabel("x")
plt.ylabel("$d^ky/dx^k$")
plt.title("ode00 analytical solutions")
plt.legend();
```
