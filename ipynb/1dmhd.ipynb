```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the run parameters.
import parameters as p
```

```python
eq_name =p.eq_name
optimizer_name = p.optimizer_name
learning_rate = p.learning_rate
H = p.H
nx_train = p.nx_train
nt_train = p.nt_train
n_train = nx_train*nt_train
```

```python
# Load the loss function histories.
losses      = np.loadtxt(os.path.join(eq_name, "losses.dat"))
losses_Bx   = np.loadtxt(os.path.join(eq_name, "losses_Bx.dat"))
losses_By   = np.loadtxt(os.path.join(eq_name, "losses_By.dat"))
losses_Bz   = np.loadtxt(os.path.join(eq_name, "losses_Bz.dat"))
losses_P    = np.loadtxt(os.path.join(eq_name, "losses_P.dat"))
losses_rho  = np.loadtxt(os.path.join(eq_name, "losses_rho.dat"))
losses_vx   = np.loadtxt(os.path.join(eq_name, "losses_vx.dat"))
losses_vy   = np.loadtxt(os.path.join(eq_name, "losses_vy.dat"))
losses_vz   = np.loadtxt(os.path.join(eq_name, "losses_vz.dat"))
```

```python
# Format the axis labels.
x_train = np.linspace(0, 1, nx_train)
t_train = np.linspace(0, 1, nt_train)
x_labels = ["%.1f" % x for x in x_train]
t_labels = ["%.1f" % t for t in t_train]
t_labels.reverse()
```

```python
# Plot the loss function histories.
plt.semilogy(losses,     label="L (total)")
plt.semilogy(losses_Bx,  label="L ($B_x$)")
plt.semilogy(losses_By,  label="L ($B_y$)")
plt.semilogy(losses_Bz,  label="L ($B_z$)")
plt.semilogy(losses_P,   label="L ($P$)")
plt.semilogy(losses_rho, label="L ($\\rho$)")
plt.semilogy(losses_vx,  label="L ($v_x$)")
plt.semilogy(losses_vy,  label="L ($v_y$)")
plt.semilogy(losses_vz,  label="L ($v_z$)")
plt.xlabel("Epoch")
plt.ylabel("Loss function")
plt.legend()
plt.grid()
plt.title("Loss function evolution for %s\n%s optimizer, $\eta$=%s, H=%s, $n_{train}$=%s" %
          (p.eq_name, p.optimizer_name, p.learning_rate, p.H, n_train))
plt.show()
```

```python
Bx_train = np.loadtxt(os.path.join(eq_name, "Bx_train.dat")).reshape(nx_train, nt_train)

# For a Seaborn heat map, reshape as (n_x, n_y), then transpose, then flip.
Z = np.flip(Bx_train.T, axis=0)
ax = sns.heatmap(Z, xticklabels=x_labels, yticklabels=t_labels)
ax.set_title("$B_x$")
ax.set_xlabel("x")
ax.set_ylabel("t")
plt.show()
```

```python
plt.plot(x_train, Bx_train[:, 0])
plt.plot(x_train, Bx_train[:, 10], label="t=1")
plt.plot(x_train, Bx_train[:, 20], label="t=2")
plt.plot(x_train, Bx_train[:, 30], label="t=3")
plt.plot(x_train, Bx_train[:, 40], label="t=4")
plt.plot(x_train, Bx_train[:, 50], label="t=5")
plt.plot(x_train, Bx_train[:, 60], label="t=6")
plt.plot(x_train, Bx_train[:, 70], label="t=7")
plt.plot(x_train, Bx_train[:, 80], label="t=8")
plt.plot(x_train, Bx_train[:, 90], label="t=9")
plt.plot(x_train, Bx_train[:, 100], label="t=10")
plt.legend()
plt.title("$B_x$")
plt.xlabel("x")
plt.show()
```
