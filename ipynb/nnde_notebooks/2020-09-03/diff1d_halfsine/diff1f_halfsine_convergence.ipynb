```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rmserr_trained_solution = np.loadtxt('rmserr_trained_solution.dat')
```

```python
rmserr_trained_gradient = np.loadtxt('rmserr_trained_gradient.dat')
```

```python
rmserr_trained_laplacian = np.loadtxt('rmserr_trained_laplacian.dat')
```

```python
rmserr_random_solution = np.loadtxt('rmserr_random_solution.dat')
```

```python
rmserr_random_gradient = np.loadtxt('rmserr_random_gradient.dat')
```

```python
rmserr_random_laplacian = np.loadtxt('rmserr_random_laplacian.dat')
```

```python
iterations = np.loadtxt('iterations.dat')
```

```python
training_time = np.loadtxt('training_time.dat')
```

```python
H = np.array(range(5, 11))
```

```python
nxnt = np.array(range(5, 12))
```

```python
for (h, row) in zip(H, rmserr_trained_solution):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('rms err')
plt.legend()
plt.title('Error in trained solution')
plt.show()
```

```python
for (h, row) in zip(H, rmserr_trained_gradient):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('H')
plt.legend()
plt.title('Error in trained gradient')
plt.show()
```

```python
for (h, row) in zip(H, rmserr_trained_laplacian):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('rms err')
plt.legend()
plt.title('Error in trained Laplacian')
plt.show()
```

```python
for (h, row) in zip(H, rmserr_random_solution):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('rms err')
plt.legend()
plt.title('Error in random solution')
plt.show()
```

```python
for (h, row) in zip(H, rmserr_random_gradient):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('rms err')
plt.legend()
plt.title('Error in random gradient')
plt.show()
```

```python
for (h, row) in zip(H, rmserr_random_laplacian):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('rms err')
plt.legend()
plt.title('Error in random Laplacian')
plt.show()
```

```python
for (h, row) in zip(H, iterations):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('Iterations')
plt.legend()
plt.title('Iterations to convergence')
plt.show()
```

```python
for (h, row) in zip(H, training_time):
    plt.plot(nxnt, row, label='H=%s' % h)
plt.xlabel('nx=nt')
plt.ylabel('Training time (s)')
plt.legend()
plt.title('Training time')
plt.show()
```
